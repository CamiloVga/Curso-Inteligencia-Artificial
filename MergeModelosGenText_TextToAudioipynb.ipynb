```python
import os
import nltk
import numpy as np
from transformers import pipeline
from bark import SAMPLE_RATE, generate_audio, preload_models
from IPython.display import Audio

# Descargar recursos necesarios de NLTK
nltk.download('punkt')

# Precargar modelos de Bark
preload_models()

# Configurar el modelo de generación de texto
model_name = "anslin-raj/Llama-2-7b-chat-hf-8-bit"
task = "text-generation"
model_gen = pipeline(task, model=model_name)

```

```python
def generate_podcast_script(topic):
    prompt = f"""Crea un diálogo entre un hombre y una mujer sobre {topic}.
    Alterna sus respuestas claramente con las etiquetas 'WOMAN:' y 'MAN:'.
    Manténlo conciso, alrededor de 5-6 intercambios en total."""

    response = model_gen(prompt, max_length=200, num_return_sequences=1)
    script = response[0]['generated_text']

    if "WOMAN:" not in script or "MAN:" not in script:
        raise ValueError("El script generado no contiene los marcadores esperados 'WOMAN:' o 'MAN:'.")

    return script

def process_script(script):
    # Dividir el script por líneas
    script_lines = script.split('\n')
    processed_lines = []
    for line in script_lines:
        line = line.strip()
        if line.startswith('WOMAN:') or line.startswith('MAN:'):
            processed_lines.append(line)
    return processed_lines  # Devolver una lista de líneas procesadas

def generate_audio_from_script(script_lines):
    silence = np.zeros(int(0.15 * SAMPLE_RATE))
    pieces = []

    for line in script_lines:
        if line.startswith('WOMAN:'):
            speaker = "v2/es_speaker_6"
            text = line.replace('WOMAN:', '').strip()
        elif line.startswith('MAN:'):
            speaker = "v2/es_speaker_8"
            text = line.replace('MAN:', '').strip()
        else:
            continue

        if text:
            try:
                audio_array = generate_audio(text, history_prompt=speaker)
                pieces.append(audio_array)
                pieces.append(silence.copy())
            except Exception as e:
                print(f"Error al generar audio para el texto: '{text}'. Detalles: {e}")

    if not pieces:
        raise ValueError("No se generaron piezas de audio. Verifica el formato del script.")

    return np.concatenate(pieces)

# Ejecución Principal
if __name__ == "__main__":
    topic = "la inflación en la economía actual"
    try:
        raw_script = generate_podcast_script(topic)
        print("Script Generado en Crudo:")
        print(raw_script)

        processed_script = process_script(raw_script)
        print("\nScript Procesado:")
        for line in processed_script:
            print(line)

        audio_array = generate_audio_from_script(processed_script)
        display(Audio(audio_array, rate=SAMPLE_RATE))
    except ValueError as e:
        print(f"Error: {e}")
```
