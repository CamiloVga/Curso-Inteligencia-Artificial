```python
from diffusers import DiffusionPipeline

pipeline = DiffusionPipeline.from_pretrained("stabilityai/sdxl-turbo")
pipeline = pipeline.to("cuda")
```

```python
prompt="a man running"
image = pipeline(prompt, num_inference_steps=100, num_images_per_prompt=1).images[0]

# Mostrar o guardar la imagen
image.save("output.png")  # Para guardar la imagen
```
